```python
import matplotlib.pyplot as plt
import itertools as it

def isortSteps(aux):
    steps = 0
    comp = 0
    swaps = 0
    v = []
    for e in aux: 
        v.append(e)
    for i in xrange(1,len(v)):
        for j in xrange(len(v)-i):    
            comp += 1
            if(v[j] > v[j+1]):
                aux = v[j];
                v[j] = v[j + 1];
                v[j + 1] = aux;
                swaps += 1;
    return [swaps,comp,steps]

"""Main"""
for i in range(11)[2:]:
    j = 0
    pComp = 0
    pSteps = 0
    pSwaps = 0
    comp = {}
    steps = {}
    swaps = {}
    a = list(it.permutations(xrange(i)))
    for e in a:
        r = isortSteps(e)
        j += 1
        pSwaps += r[0]
        pComp += r[1]
        pSteps += r[2]
        if(r[0] in swaps):
            swaps[r[0]] += 1
        else:
            swaps[r[0]] = 1
        if(r[1] in comp):
            comp[r[1]] += 1
        else:
            comp[r[1]] = 1
        if(r[2] in steps):
            steps[r[2]] += 1
        else:
            steps[r[2]] = 1
    print i
    j = float(j)
    print "Promedio comparaciones:", pComp/j
    print "Promedio swaps:", pSwaps/j
    print "Promedio pasos", pSteps/j
    print "Distribución comparaciones:", comp
    print "Distribución swaps:", swaps
    print "Distribución pasos:", steps
    aux1 = steps.keys()
    aux2 = []
    aux1.sort()
    for e in aux1:
        aux2.append(steps[e])
    plt.figure(figsize=(20,5))
    plt.bar(range(len(aux2)), aux2, align='center')
    plt.xticks(range(len(aux1)), aux1)
    plt.show()
```
