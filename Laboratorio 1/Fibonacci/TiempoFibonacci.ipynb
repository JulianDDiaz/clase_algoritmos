```python
import time
import matplotlib.pyplot as pl
arr = []
for i in xrange(50000):
    n1 = 1
    n2 = 0
    start_time = time.time()
    for j in xrange(i):
        a = n1+n2
        n2 = n1
        n1 = a
    arr.append((time.time() - start_time))
pl.plot(range(50000),arr)
```
